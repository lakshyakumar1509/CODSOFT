```python
label_encoder = LabelEncoder()
iris_data['species'] = label_encoder.fit_transform(iris_data['species'])
```

```python
correlation_matrix = iris_data.corr()
plt.figure(figsize=(5, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
X_features = iris_data.drop(columns=['species'])
y_target = iris_data['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_features, y_target, test_size=0.30, random_state=42)
```

```python
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
logistic_accuracy = logistic_model.score(X_test, y_test) * 100
print(f'Logistic Regression Accuracy: {logistic_accuracy:.2f}%')
```

```python
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)
knn_accuracy = knn_model.score(X_test, y_test) * 100
print(f'K-Nearest Neighbors Accuracy: {knn_accuracy:.2f}%')
```

```python
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(X_train, y_train)
decision_tree_accuracy = decision_tree_model.score(X_test, y_test) * 100
print(f'Decision Tree Accuracy: {decision_tree_accuracy:.2f}%')
```
